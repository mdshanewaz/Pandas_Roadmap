```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first' : ['Shah', 'Shah', 'Kawser', 'Adam', np.nan, None, 'NA'],
    'last' : ['Newaz', 'Waliullah', 'Rana', 'Doe', np.nan, np.nan, 'Missing'],
    'email' : ['demo1@mail.com', 'demo2@mail.com', 'demo3@mail.com', None, np.nan, 'annnyem@mail.com', 'NA'],
    'age' : [23, 34, 45, 29, None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
```
