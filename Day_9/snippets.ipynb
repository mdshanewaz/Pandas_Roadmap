```python
import pandas as pd
```

```python
people = {
    'first' : ['Shah', 'Shah', 'Kawser', 'Adam'],
    'last' : ['Newaz', 'Waliullah', 'Rana', 'Doe'],
    'email' : ['demo1@mail.com', 'demo2@mail.com', 'demo3@mail.com', 'ado@mail.com'],
    'age' : [23, 34, 45, 29]
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.sort_values(by='last')
```

```python
df.sort_values(by='last', ascending=False)
```

```python
df.sort_values(by=['last', 'first'], ascending=False)
```

```python
df.sort_values(by=['last', 'first'], ascending=[False, True])
```
